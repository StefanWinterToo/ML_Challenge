```python
import numpy
import os
import cv2
from cv2 import Canny
from matplotlib import pyplot as plt
```

```python
def load_images_from_folder(folder):
    images = []
    for file in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, file), cv2.IMREAD_GRAYSCALE)
        if img is not None:
            images.append(img)
        return images

path = "../../data/sample_dataset/"
images = load_images_from_folder(path)
```

```python
def plot_n_images(data, i = 1):
    d = data[:i]
    for i in range(len(d)):
        plt.imshow(d[i].reshape(28, 200), cmap="gray")
        plt.show()
```

```python
from skimage.feature import canny
from scipy import ndimage as ndi
dataset = dataset[:10] #Subset of dataset
```

```python
edges = canny(dataset[0]/255., sigma = 3)
```

```python
fill_data = ndi.binary_fill_holes(edges)
```

```python
for i in range(len(dataset)):
    edges = canny(dataset[i]/255., sigma = 1.3)
    fill_data = ndi.binary_fill_holes(edges)

    # Edges
    plt.imshow(fill_data, cmap="gray")
    plt.show()

    # Original
    plt.imshow(dataset[i], cmap = "gray")
    plt.show()
    print()
    print("--------")
```

```python
blurred = cv2.GaussianBlur(dataset[0], (7, 7), 5)
plt.imshow(blurred, cmap="gray")
plt.show()
```

```python
edges = canny(blurred/255., sigma = 0.7)
fill_data = ndi.binary_fill_holes(edges)

# Edges
plt.imshow(fill_data, cmap="gray")
plt.show()

# Original
plt.imshow(blurred, cmap = "gray")
plt.show()
print()
print("--------")
```
